```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelBinarizer
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.feature_selection import f_classif, SelectKBest
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score, roc_auc_score, roc_curve
from sklearn.metrics import mean_squared_error, mean_absolute_error, classification_report
```

```python
# 'Class' feature'un da bulunan string yapılarını numeric halde yazma
_class = {'DERMASON':0, 'SIRA':1, 'SEKER':2, 'HOROZ':3, 'CALI':4, 'BARBUNYA':5, 'BOMBAY':6}
veri['Class'] = veri['Class'].replace(_class)

# Girdi ve Çıktı olan özelliklerini ayırma
X = veri.drop(columns='Class')
y = veri['Class']
```

```python
class_values = veri['Class'].values
print(class_values, "\n")
class_values_binary = label_binary.fit_transform(class_values)
print(class_values_binary, "\n")
class_counts = class_values_binary.sum(axis=0)
print(class_counts)
```

```python
# Pie Charts
class_values = veri['Class'].values
label_binary = LabelBinarizer()
mylabels = ["DERMASON", "SIRA", "SEKER", "HOROZ", "CALI", "BARBUNYA", "BOMBAY"]
class_values_binary = label_binary.fit_transform(class_values)
class_counts = class_values_binary.sum(axis=0)
plt.pie(class_counts, labels=mylabels, autopct='%1.1f%%', shadow = True)
plt.show()
```

```python
# K-fold
kf = KFold(n_splits=10, random_state=44, shuffle=True)

for train_index, test_index in kf.split(X):
    X_train, y_train = X.iloc[train_index], y.iloc[train_index]
    X_test, y_test = X.iloc[test_index], y.iloc[test_index]

fs = SelectKBest(f_classif, k=10)
fs.fit(X_train, y_train)
X_train_selected = fs.transform(X_train)
X_test_selected = fs.transform(X_test)
```

```python
# Logistic Regression

lrModel = LogisticRegression(C=1, solver='newton-cg', class_weight='balanced', multi_class='multinomial',
                            fit_intercept=True, max_iter=100, random_state=44)
lrModel.fit(X_train_selected, y_train)

# Test ve Eğitim verileri
train_accuracy = lrModel.score(X_train_selected, y_train)
print("Eğitim verisinin doğruluğu : %.5f " %train_accuracy)
test_accuracy = lrModel.score(X_test_selected, y_test)
print("Test verisinin doğruluğu   : %.5f \n" %test_accuracy)

# 10fold Cross-Validation
scores = cross_val_score(lrModel, X_test_selected, y_test, cv=kf)
print("LR Modelinin CV performansı : ", scores)
print("LR Modelinin CV performansının ortalaması : %.5f \n" %scores.mean())

# Confusion Matrix
y_pred = lrModel.predict(X_test_selected)
Confi_Matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(Confi_Matrix/np.sum(Confi_Matrix), annot=True, center=True, fmt=".2g")
plt.show()
print("Confusion Matrix : \n", Confi_Matrix)

# Performans Metrikleri
accuracy_lr = accuracy_score(y_test, y_pred)
recall_lr = recall_score(y_test, y_pred,average='macro')
precision_lr = precision_score(y_test, y_pred, average = 'macro')
f1_lr = f1_score(y_test, y_pred, average = 'macro')

print("\n\nLR Modelin Accuracy Değeri  : %.5f" % accuracy_lr)
print("LR Modelin Recall Değeri    : %.5f" % recall_lr)
print("LR Modelin Precision Değeri : %.5f" % precision_lr)
print("LR Modelin F1 Skoru         : %.5f" % f1_lr)

# Hata Metrikleri
mae_lr = mean_absolute_error(y_test, y_pred)
mse_lr = mean_squared_error(y_test, y_pred)

print("\nLR Modelin MAE değeri : %.5f" % mae_lr)
print("LR Modelin MSE değeri : %.5f" % mse_lr)

# Grafikler
metrikler = pd.DataFrame({"Metrikler": ["Accuracy", "Precision", "Recall", "F1", "MAE", "MSE"], 
                          "Değerler": [accuracy_lr, recall_lr, precision_lr, f1_lr, mae_lr, mse_lr]})
vr_ac = pd.DataFrame({"Doğruluk": ["Eğitim_Verisi", "Test_Verisi"], 
                          "Değerler": [train_accuracy, test_accuracy]})

fig = plt.figure(figsize=(15, 5))
fig.add_subplot(1, 2, 1)
sns.barplot(x="Metrikler", y="Değerler", data=metrikler)
fig.add_subplot(1, 2, 2)
sns.barplot(x="Doğruluk", y="Değerler", data=vr_ac)
```
